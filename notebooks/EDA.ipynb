```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd

# Define the file path where your data is located
file_path = '/Users/rajhabib/customer-churn-prediction/data/preprocessed_data.csv'

# Reload the data into a DataFrame
df = pd.read_csv(file_path)
# Define custom color palettes or colors for each plot
color_palettes = {
    'gender_churn': 'Set2',
    'contract_churn': 'tab10',
    'payment_distribution': 'viridis',
    'payment_distribution_churn': 'magma',
    'churn_internet_service': 'Pastel1',
    'dependents': 'tab20',
    'churn_partners': 'Paired',
    'churn_senior_citizens': 'cubehelix',
    'churn_online_security': 'Spectral',
    'churn_paperless_billing': 'Set3',
    'churn_tech_support': 'Dark2',
    'churn_phone_service': 'tab20c',
    'monthly_charges_distribution': 'rocket',
    'tenure_vs_churn': 'plasma'
}

def annotate_percentages(ax):
    total = len(df)  # Total number of samples
    for p in ax.patches:
        height = p.get_height()
        if height > 0:  # Skip annotation if height is zero
            ax.annotate(f'{height / total * 100:.1f}%', 
                        (p.get_x() + p.get_width() / 2., height), 
                        ha='center', va='center', 
                        xytext=(0, 5), 
                        textcoords='offset points')

        
# Plot gender and churn distribution
plt.figure(figsize=(8, 5))
ax = sns.countplot(x='gender_Male', hue='Churn', data=df, palette=color_palettes['gender_churn'])
plt.title('Gender and Churn Distribution')
plt.xlabel('Gender')
plt.xticks([0, 1], ['Female', 'Male'])
annotate_percentages(ax) 
plt.show()

# Plot customer contract distribution with churn or no churn
plt.figure(figsize=(8, 5))
ax = sns.countplot(x='Contract_One year', hue='Churn', data=df, palette=color_palettes['contract_churn'])
plt.title('Customer Contract Distribution with Churn or No Churn')
plt.xlabel('Contract Type')
annotate_percentages(ax) 
plt.show()

# Plot customer payment distribution
plt.figure(figsize=(8, 5))
sns.histplot(x='MonthlyCharges', hue='Churn', data=df, bins=30, kde=True, palette=color_palettes['payment_distribution'])
plt.title('Customer Payment Distribution')
plt.xlabel('Monthly Charges')
plt.show()

# Plot customer payment distribution with churn or no churn
plt.figure(figsize=(8, 5))
sns.histplot(x='MonthlyCharges', hue='Churn', data=df, bins=30, kde=True, multiple='stack', palette=color_palettes['payment_distribution_churn'])
plt.title('Customer Payment Distribution with Churn or No Churn')
plt.xlabel('Monthly Charges')
plt.show()

# Plot churn rates based on different internet services
plt.figure(figsize=(10, 6))
sns.countplot(x='InternetService', hue='Churn', data=df, palette='Set2', dodge=True, order=['Fiber optic', 'DSL', 'No'])

# Customize the plot
plt.title('Churn Rates by Internet Service')
plt.xlabel('Internet Service')
plt.ylabel('Count')
plt.legend(title='Churn', loc='upper right', labels=['No Churn', 'Churn'])

# Show plot
plt.show()


# Plot dependents distribution
plt.figure(figsize=(8, 5))
ax = sns.countplot(x='Dependents_Yes', hue='Churn', data=df, palette=color_palettes['dependents'])
plt.title('Dependents Distribution with Churn')
plt.xlabel('Dependents (Yes/No)')
annotate_percentages(ax) 
plt.show()

# Plot churn based on partners
plt.figure(figsize=(8, 5))
ax = sns.countplot(x='Partner_Yes', hue='Churn', data=df, palette=color_palettes['churn_partners'])
plt.title('Churn Based on Partners')
plt.xlabel('Partner (Yes/No)')
annotate_percentages(ax) 
plt.show()

# Plot churn based on senior citizens
plt.figure(figsize=(8, 5))
ax = sns.countplot(x='SeniorCitizen', hue='Churn', data=df, palette=color_palettes['churn_senior_citizens'])
plt.title('Churn Based on Senior Citizens')
plt.xlabel('Senior Citizen (Yes/No)')
annotate_percentages(ax) 
plt.show()

# Plot churn with online security
plt.figure(figsize=(8, 5))
ax = sns.countplot(x='OnlineSecurity', hue='Churn', data=df, palette=color_palettes['churn_online_security'])
plt.title('Churn with Online Security')
plt.xlabel('Online Security (Yes/No)')
annotate_percentages(ax) 
plt.show()

# Plot churn with paperless billing
plt.figure(figsize=(8, 5))
ax = sns.countplot(x='PaperlessBilling', hue='Churn', data=df, palette=color_palettes['churn_paperless_billing'])
plt.title('Churn with Paperless Billing')
plt.xlabel('Paperless Billing (Yes/No)')
annotate_percentages(ax) 
plt.show()

# Plot churn with tech support
plt.figure(figsize=(8, 5))
ax = sns.countplot(x='TechSupport', hue='Churn', data=df, palette=color_palettes['churn_tech_support'])
plt.title('Churn with Tech Support')
plt.xlabel('Tech Support (Yes/No)')
annotate_percentages(ax) 
plt.show()

# Plot churn with phone service
plt.figure(figsize=(8, 5))
ax = sns.countplot(x='PhoneService', hue='Churn', data=df, palette=color_palettes['churn_phone_service'])
plt.title('Churn with Phone Service')
plt.xlabel('Phone Service (Yes/No)')
annotate_percentages(ax) 
plt.show()

# Plot distribution of monthly charges by churn
plt.figure(figsize=(10, 6))
sns.boxplot(x='Churn', y='MonthlyCharges', data=df, palette=color_palettes['monthly_charges_distribution'])
plt.title('Distribution of Monthly Charges by Churn')
plt.xlabel('Churn')
plt.ylabel('Monthly Charges')
plt.show()

# Plot tenure vs churn
plt.figure(figsize=(10, 6))
sns.histplot(x='tenure', hue='Churn', data=df, bins=30, kde=True, palette=color_palettes['tenure_vs_churn'])
plt.title('Tenure vs Churn')
plt.xlabel('Tenure (months)')
plt.show()
```
